```python
df2 = pd.read_excel('test.xlsx', sheetname = 'Sheet1', 
                    header = None, 
                    index_col = 0, 
                    skiprows = 1)
```

```python
# read all sheets from excel with multiple sheets

xls = pd.ExcelFile('test.xlsx')
xls.sheet_names # the list of all the sheets
sheet1 = xls.parse[0] # the first sheet
sheet2 = xls.parse['sheetname'] # a specific sheet by name
```

```python
# write to an Excel file
df1.to_excel('result.xlsx', 
            index = True # by default
            )
```

```python
# write to multiple sheet in same excel file
writer = pd.ExcelWriter('result2.xlsx', engine = 'xlsxwriter')
# writer = pd.ExcelWriter('result2.xlsx', engine='openpyxl')

df1.to_excel(writer, sheet_name = 'df1')
df2.to_excel(writer, sheet_name = 'df2')
writer.save()
```

## Clean Data

```python
# deal with NaN
# drop None
df.dropna(axis = 0, # by default
          how = 'any', # by default, can be "all"
          thresh = None, # by default
          subset = None, # by default
          inplace = False # by default. if it is True, it will change the origianl data and return nothing
         )
# fill None
df.fillna('Unknown')
df['Label'].fillna('Unknown')
```

```python
# drop columns
df.drop(['Label'],
       axis = 1
       )
```

```python
df.drop_duplicates(
    keep = 'first' # by default
)
```
